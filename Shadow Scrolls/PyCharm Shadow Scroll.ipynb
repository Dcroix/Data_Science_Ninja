## <font face = 'Impact' color = '#C87620' > Shadow Step (Navigation) <font/>
- Ctrl + N / Cmd + O → Summon any class instantly
- Ctrl + Shift + N / Cmd + Shift + O → Blink to any file
- Ctrl + Alt + Shift + N / Cmd + Option + O → Target method like a throwing star

## <font face = 'Impact' color = '#C87620' > Silent Strike (Editing) <font/>
- Ctrl + D / Cmd + D → Duplicate line with ninja precision
- Ctrl + Y / Cmd + Backspace → Delete line—vanish without trace
- Ctrl + Shift + ↑/↓ / Cmd + Shift + ↑/↓ → Shift lines like a shadow sliding
- Ctrl + / / Cmd + / → Comment/uncomment like a stealth spell

## <font face = 'Impact' color = '#C87620' > Smoke Bomb (Refactoring) <font/>
- Shift + F6 → Rename without leaving a footprint
- Ctrl + Alt + V / Cmd + Option + V → Extract variable like a ninja pulls secrets
- Ctrl + Alt + M / Cmd + Option + M → Extract method—split code like bamboo

## <font face = 'Impact' color = '#C87620' > Lightning Strike (Running & Debugging) <font/>
- Shift + F10 / Ctrl + R → Launch the mission (run code)
- Shift + F9 / Cmd + Option + R → Enter the debug dojo
- F8 → Step over like a silent footstep
- F7 → Step into—penetrate the shadows of your code

## <font face = 'Impact' color = '#C87620' > Mystic Vision (Search & Replace) <font/>
- Ctrl + F / Cmd + F → Scan your scrolls (search in file)
- Ctrl + R / Cmd + R → Change destiny (replace in file)
- Ctrl + Shift + F / Cmd + Shift + F → Seek across the kingdom (global search)

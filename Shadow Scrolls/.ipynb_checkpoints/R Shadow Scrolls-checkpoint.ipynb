## Script Shortcuts

| Shortcut (Windows/Linux) | Shortcut (Mac)        | Notes                               |
| ------------------------ | --------------------- | ----------------------------------- |
| `Ctrl + Shift + N`       | `Cmd + Shift + N`     | Open a blank R script               |
| `Ctrl + O`               | `Cmd + O`             | Navigate to an existing file        |
| `Ctrl + S`               | `Cmd + S`             | Save current script                 |
| `Ctrl + Enter`           | `Cmd + Enter`         | Execute code instantly              |
| `Ctrl + Shift + Enter`   | `Cmd + Shift + Enter` | Moves cursor to next line           |
| `Ctrl + Shift + C`       | `Cmd + Shift + C`     | Toggles `#` on selected lines       |
| `Tab` / `Shift + Tab`    | Same                  | Format code neatly                  |
| `Ctrl + .`               | `Cmd + .`             | Quick jump to any script in project |

## Console Shorcuts

| Shortcut (Windows/Linux) | Shortcut (Mac)   | Notes                                    |
| ------------------------ | ---------------- | ---------------------------------------- |
| `Ctrl + L`               | `Cmd + L`        | Wipe previous outputs                    |
| `Esc`                    | `Esc`            | Stop running code (emergency ninja move) |
| `Ctrl + Up Arrow`        | `Cmd + Up Arrow` | Scroll through previous commands         |
| `Ctrl + Enter`           | `Cmd + Enter`    | Same as in editor                        |

## Navigation & Search

| Action              | Shortcut                            | Notes                              |
| ------------------- | ----------------------------------- | ---------------------------------- |
| Go to File / Symbol | `Ctrl + .`                          | Jump anywhere in your project      |
| Find / Replace      | `Ctrl + F` / `Ctrl + Shift + F`     | Search in current file / all files |
| Go to line          | `Ctrl + G`                          | Direct jump to a specific line     |
| Next / Previous tab | `Ctrl + Tab` / `Ctrl + Shift + Tab` | Move through open scripts          |

## Plots or Panes Shortcuts

| Action            | Shortcut               | Notes                              |
| ----------------- | ---------------------- | ---------------------------------- |
| Zoom plot         | `Alt + Shift + Z`      | Focus on the plot pane             |
| Clear plots       | `Ctrl + Shift + F10`   | Reset plots in session             |
| Switch pane focus | `Ctrl + 1 / 2 / 3 / 4` | Editor, Console, Environment, Help |

## Project / Session Shortcuts

| Action            | Shortcut             | Notes                              |
| ----------------- | -------------------- | ---------------------------------- |
| Restart R session | `Ctrl + Shift + F10` | Reset R without closing RStudio    |
| Knit document     | `Ctrl + Shift + K`   | Render R Markdown to HTML/PDF/Word |
| Open project      | `Ctrl + Shift + O`   | Jump into a new project            |
| Save workspace    | `Ctrl + Shift + S`   | Save current environment           |

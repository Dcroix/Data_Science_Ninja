## File & Project Management
- Ctrl + N         → Open a new Python file          
- Ctrl + O         → Navigate to an existing script   
- Ctrl + S         → Save current file                
- Ctrl + Shift + S → Save under a new name            
- Ctrl + W         → Close active tab               
- Ctrl + Tab       → Cycle between open files         

## Code and Editing Navigation

- F5 →	Execute the current file
- F9 →	Useful for testing snippets
- Ctrl + 1 → Toggles # for selected lines
- Tab / Shift + Tab → Align code properly
- Ctrl + Click → Jump to function/class definition
- Ctrl + F → Search within file
- Ctrl + H → Search & replace in file

## Consol / IPython 

- Ctrl + 2 → Switch to console window
- F9 → Runs highlighted code in console
- Ctrl + L → Wipe previous outputs
- Ctrl + . → Stop a running script

## Debugging Mode

- F12 → Click gutter or use key
- Ctrl + F5 → Start debugger without stopping on first line
- F11 → Go inside function calls
- F10 → Run line without entering function
- Ctrl + F12 → Resume until next breakpoint
- Shift + F12 → Stop debugging

## Editor Navigation

- Ctrl + G → Jump directly to a line number
- Ctrl + Alt + - → Backtrack edits
- Ctrl + Alt + + → Forward in edits
- Ctrl + Shift + O → See functions/classes in file

## Other Spyder Tricks

- F11 → Focus on code
- Ctrl + Shift + O → View multiple files side-by-side
- Ctrl + Alt + Shift + P → Customize your setup
- Ctrl + - → Minimize all functions/classes
- Ctrl + Shift + + → Restore all folds

This is helpful for individuals learning the jupyter notebook application. It is a quick-reference guide to the most essential command and edit mode shortcuts in Jupyter Notebook, helping you run cells, navigate, edit, and manage notebooks more efficiently without using the mouse.

## Run and Navigation

- Shift + Enter → Run cell and move to next  
- Ctrl + Enter → Run cell and stay
- Alt + Enter → Run cell and insert new below
- Up / Down → Move between cells

## Cell Management

- A → Insert cell Above
- B → Insert cell Below
- DD → Delete cell
- Z → Undo delete cell
- C → Copy cell
- X → Cut cell
- V → Paste below
- Shift + V → Paste above

# Change Cell Type

- Y → Code cell
- M → Markdown cell
- R → Raw cell

# Other quick keys

- H → Show all shortcuts
- I, I → Interrupt kernel
- 0, 0 → Restart kernel
- Shift + M → Merge selected cells
- Ctrl + S → Save notebook
